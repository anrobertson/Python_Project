```python
from faker import Faker     
import numpy as np
import pandas as pd

# Creating fake companies
companies= ['Hershe','Campbull','Costgo','Walsmart','Procter & Deal','Kroft','General Malls','Betty Socker','Chiquota','Nosson Foods','Del Marte','Folders']

fake = Faker() 

def generate_fake_address():
    full_address = fake.address()
    
    # Remove the zip code, state, and city from the address
    parts = full_address.split('\n')
    address_line1 = parts[0]
    address_line2 = '\n'.join(parts[1:])
    
    return address_line1, address_line2

def create_data(x): 
    b_user = {} 
    for i in range(x): 
        b_user[i] = {} 
        b_user[i]['UMI'] = fake.random_int(min=2030000000000, max=2039999999999)
        b_user[i]['Sender Level2'] = fake.random_int(min=4000000000, max=4000999999)
        b_user[i]['Sender Level2 Name'] = fake.random_element(companies)
        b_user[i]['Sender Level1'] = fake.random_int(min=4000000000, max=4000999999)
        b_user[i]['Sender Level1 Name'] = fake.company()
        b_user[i]['Sender'] = fake.random_int(min=4000000000, max=4000999999)
        b_user[i]['Sender Name'] = fake.company()
        b_user[i]['Sender Ext Code'] = fake.random_int(min=4000234512, max=79999999999)
        b_user[i]['Sender GLID City'] = fake.city()
        b_user[i]['Sender Audit Date'] = fake.date_this_year()
        b_user[i]['Sender Channel'] = 'Emitter'
        b_user[i]['Sender_Owner'] = fake.name()
        b_user[i]['Sender Account Executive'] = '-'
        b_user[i]['Mvmt Receiver Name'] = fake.company()
        
        # Generate fake address without zip code, state, or city
        address_line1, _ = generate_fake_address()  # Remove the unused address_line2
        b_user[i]['Mvmt Receiver Address'] = address_line1
        
        state = fake.state()
        city = fake.city()
        zipcode = fake.postcode()
        
        b_user[i]['Mvmt Receiver City'] = city
        b_user[i]['Mvmt Receiver Region'] = state
        b_user[i]['Mvmt Receiver Zipcode'] = zipcode
        
        b_user[i]['DOD'] = fake.date_this_year()
        b_user[i]['Unk_DON'] = fake.date_this_year()
        b_user[i]['Qty'] = fake.random_int(min=1, max=40)
        
        # Generate numbers in the format of "10-20," "30-40," etc. for the age bucket
        lower_bound = fake.random_int(min=1, max=9) * 10
        upper_bound = lower_bound + 10
        b_user[i]['Age Bucket'] = f"{lower_bound}-{upper_bound}"    
        
    return b_user
    
df = pd.DataFrame(create_data(50000)).transpose()
df.head(5)
```

```python
# Save the DataFrame as a CSV file
csv_file_name = 'mock_data.csv'
df.to_csv(csv_file_name, index=False, float_format='%f')  # %f ensures full precision

print(f"CSV file '{csv_file_name}' has been created.")
```
